# pyCellModeller API Flow (CPU)

Minimal local exploration notebook using the same flow as `examples/scripts/deterministic_cpu_summary.py`.

```python
from pycellmodeller import CellProgram, Simulation, SimulationConfig, TorchBacteriumimport torch
```

```python
class NoOpProgram(CellProgram):
    pass

config = SimulationConfig(device="cpu", dt=0.1, seed=1)
simulation = Simulation(
    config=config,
    biophysics=TorchBacterium(growth_rate=0.0, division_length=10.0),
    program=NoOpProgram(),
)
state0 = simulation.initialize()
simulation.add_cell(position=torch.tensor([0.0, 0.0]), velocity=torch.tensor([0.25, 0.0]))
simulation.add_cell(position=torch.tensor([1.0, 0.0]), velocity=torch.tensor([0.0, 0.25]))
state1 = simulation.step()
summary = {
    "step_index": state1.step_index,
    "time": round(state1.time, 6),
    "position_sum": [round(v, 6) for v in state1.positions.sum(dim=0).tolist()],
    "velocity_sum": [round(v, 6) for v in state1.velocities.sum(dim=0).tolist()],
}
summary
```
