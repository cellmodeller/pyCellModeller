# pyCellModeller API Flow (CPU)

Minimal local exploration notebook using the same flow as `examples/scripts/deterministic_cpu_summary.py`.

```python
from pycellmodeller import Simulation, SimulationConfig
```

```python
config = SimulationConfig(device="cpu", dt=0.1, seed=1)
simulation = Simulation(config)
state0 = simulation.initialize_example()
state1 = simulation.step()
summary = {
    "step_index": state1.step_index,
    "time": round(state1.time, 6),
    "position_sum": [round(v, 6) for v in state1.positions.sum(dim=0).tolist()],
    "velocity_sum": [round(v, 6) for v in state1.velocities.sum(dim=0).tolist()],
}
summary
```
